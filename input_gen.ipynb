```python
def chunks(lst, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
gen = chunks(names, 30)
buses = [next(gen) for _ in range(13)]
for b in buses:
    print(b)
```

```python
def add_edges(buses, G):
    for bus in buses:
        edge_lst = []
        for n1 in bus:
            for n2 in bus:
                if n1 != n2:
                    edge_lst.append((n1, n2))
        G.add_edges_from(edge_lst)
def generate_constraints(n, buses):
    constraints = []
    for i in range(n):
        lst = [bus[i] for bus in buses[1:]]
        for name in buses[0]:
            constraints.append(lst + [name])
    return constraints
```
